```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, BatchNormalization

# Load MNIST data
(X_train, _), (_, _) = mnist.load_data()
X_train = (X_train.astype(np.float32) - 127.5) / 127.5
X_train = X_train.reshape(-1, 28 * 28)  # Reshape for autoencoder input

# Autoencoder
autoencoder = Sequential([
    Dense(512, input_dim=X_train.shape[1]),  # Adjust input dimension based on your data shape
    LeakyReLU(0.2),
    BatchNormalization(),
    Dense(256),
    LeakyReLU(0.2),
    BatchNormalization(),
    Dense(128),
    LeakyReLU(0.2),
    BatchNormalization(),
    Dense(64, activation='tanh'),  # Bottleneck layer
    Dense(128),
    LeakyReLU(0.2),
    BatchNormalization(),
    Dense(256),
    LeakyReLU(0.2),
    BatchNormalization(),
    Dense(512),
    LeakyReLU(0.2),
    BatchNormalization(),
    Dense(X_train.shape[1], activation='tanh')  # Adjust output dimension based on your data shape
])

# Compile autoencoder
autoencoder.compile(loss='mse', optimizer='adam')

# Create directory if it doesn't exist
if not os.path.exists('generated_images'):
    os.makedirs('generated_images')

# Training
epochs, batch_size = 10000, 64

for epoch in range(epochs):
    idx = np.random.randint(0, X_train.shape[0], batch_size)
    real_images = X_train[idx]

    ae_loss = autoencoder.train_on_batch(real_images, real_images)

    if epoch % 100 == 0:
        print(f"Epoch: {epoch}, AE Loss: {ae_loss}")

    if epoch % 1000 == 0:
        reconstructed_images = autoencoder.predict(X_train[:10])  # Reconstruct some images
        for i in range(10):
            plt.subplot(1, 2, 1)
            plt.imshow(real_images[i].reshape(28, 28), cmap='gray')  # Adjust reshape dimensions based on your data shape
            plt.title('Original Image')
            plt.axis('off')

            plt.subplot(1, 2, 2)
            plt.imshow(reconstructed_images[i].reshape(28, 28), cmap='gray')  # Adjust reshape dimensions based on your data shape
            plt.title('Reconstructed Image')
            plt.axis('off')

            plt.savefig(f"generated_images/autoencoder_image_{epoch}_{i}.png")
            plt.show()
            plt.close()
```
